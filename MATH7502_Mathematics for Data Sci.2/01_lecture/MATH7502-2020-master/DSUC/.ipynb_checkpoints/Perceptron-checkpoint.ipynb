### The perceptron algorithm

```julia
using Random, Distributions, Plots, LinearAlgebra; pyplot()

M = 2  #bounding box size of our world (-M,M)x(-M,M)

function dataGenerator(n = 100)
    Random.seed!(0) #set seed to make it the same on every run
    w = [1,1] #mock model
    b = 0.5  #mock model
    xPts = [[rand(Uniform(-M,M)), rand(Uniform(-M,M))] for _ in 1:n]
    label(x) = w'x + b > 0 ? +1 : -1 #one line function for the label (Y)
    y = label.(xPts) #. is the broadcast operator applying label to each xPts
    return xPts, y  #return a tuple (features and labels)
end
```

```julia
#from this point on we don't know "w" and "b"
x, y = dataGenerator(200)
xNeg = x[y .== -1]
xPos = x[y .== +1]
scatter(first.(xNeg),last.(xNeg),
        c = :blue, ms = 5, msw=0, label = "-1 label")
scatter!(first.(xPos),last.(xPos),
        c = :red, ms = 5, msw=0, label = "+1 label",legend = :topright)
```

```julia
"""
Implements the perceptron algorithm
"""
function learnPerceptron(x,y)
    n = length(x)
    w = zeros(3) #initilize w as zeros note that it now also has the bias in there
    i = 1
    history = []
    while true
        misses = 0
        for i in 1:n #try all points
            predY = w'*[1 ; x[i]] > 0 ? +1 : -1 #make a prediction based on current w
            if y[i] != predY #if bad prediction
                w = w + y[i]*[1 ; x[i]]  #key update STEP!!!!! (either add [1 x] or subtract it)
                misses += 1
                push!(history,(w,i)) #keep history for the animation (not part of algorithm)
            end
        end
        if misses == 0 
            break  #if all points are classified then stop!!!
        end
    end #end of while loop
    return w, history
end
```

```julia
extendedW ,history = learnPerceptron(x,y)
b = extendedW[1]
w = extendedW[2:end] #just 2 dim in our case
w = w/norm(w)
@show b
@show w
@show length(history)
```

$$
[w_1~~w_2]^T \cdot [-M~~x_2]^T + b == 0
$$

```julia
scatter(first.(xNeg),last.(xNeg),c = :blue, ms = 5, msw=0, label = "-1 label")
scatter!(first.(xPos),last.(xPos),c = :red, ms = 5, msw=0, label = "+1 label")
x2Low = (-b - w[1]*(-M))/w[2] #solve the above for x_2 and you get x2Low
x2High = (-b - w[1]*M)/w[2]  #Simlarly for x_1 = M (instead of -M)
plot!([-M,M],[x2Low,x2High],lw = 5, c=:black,xlim=(-M,M),ylim=(-M,M),label ="Threshold")
```

```julia
anim = Animation()
#Iterate over all of history, but also keep the last point of history for 20 iterations
for (n,h) in enumerate(vcat(history,fill(history[end],20)))
    scatter(first.(xNeg),last.(xNeg),c = :blue, ms = 5, msw =0, label = "-1 label")
    scatter!(first.(xPos),last.(xPos),c = :red, ms = 5, msw =0, label = "+1 label")
    i = h[2]
    b = h[1][1]
    w = h[1][2:end]
    x2Low = (-b - w[1]*(-M))/w[2]
    x2High = (-b - w[1]*M)/w[2]
    scatter!([first(x[i])],[last(x[i])], color = :green, ms = 10,label = "Last point")
    plot!([-M,M],[x2Low,x2High],lw = 5, c=:black,xlim=(-M,M),ylim=(-M,M),label ="Threshold")
    annotate!(0, -1.5, text("Iteration $(min(n,length(history))) out of $(length(history))", :center, :top, 20))
    frame(anim) #records the next animation frame...
end
cd(@__DIR__)
gif(anim,"perceptron.gif",fps=5)
```
