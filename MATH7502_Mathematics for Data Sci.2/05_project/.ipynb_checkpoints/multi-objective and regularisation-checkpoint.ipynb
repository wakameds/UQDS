## multi-objective

```julia
using LinearAlgebra, RDatasets, Statistics, Plots, Random, Distributions
pyplot();

# set dataset
cars = dataset("datasets", "cars")
cars = convert(Array, cars)
xVals = cars[:,1]
yVals = cars[:,2]

n = length(xVals)

function designA(xVals, t)
    powers = 0:1:t
    A = [xVals[i]^p for i in 1:n, p in powers]
end

function yHat(x, t)
    powers = 0:1:t
    (pinv(designA(xVals, t))*yVals)'*[x^p for p in powers]
end


xGrid = 0:30
p2 = scatter(xVals, yVals,legend=false, xlabel="Speed", ylabel = "Distance",
        xlims=(0, 30), ylims=(0,150))
plot!(xGrid, yHat.(xGrid,2))
title!("degree 2")

p5 = scatter(xVals, yVals,legend=false, xlabel="Speed", ylabel = "Distance",
        xlims=(0, 30), ylims=(0,150))
plot!(xGrid, yHat.(xGrid,5), label="degree = 5")
title!("degree 5")


p10 = scatter(xVals, yVals,legend=false, xlabel="Speed", ylabel = "Distance",
        xlims=(0, 30), ylims=(0,150))
plot!(xGrid, yHat.(xGrid,10))
title!("degree 10")


p20 = scatter(xVals, yVals,legend=false, xlabel="Speed", ylabel = "Distance",
        xlims=(0, 30), ylims=(0,150))
plot!(xGrid, yHat.(xGrid,20))
title!("degree 20")


plot(p2, p5, p10, p20,layout=(2,2), size=(1200,800))
```

### regularization

```julia
A  = designA(xVals, 10)

ridgeBeta(λ) = inv(A'*A + λ*I)*A'*yVals
λGrid = 0:20:600
betaEsts = ridgeBeta.(λGrid); #doing many fits, one for each \lambda
norm.(betaEsts)
```

```julia
plot(norm.(betaEsts[2:end]))
```

```julia
betaHat = ridgeBeta(600) #with lambda = 600
```

```julia
function yHat(x, t)
    powers = 0:1:t
    betaHat'*[x^p for p in powers]
end

xGrid = 1:30
yEst = yHat.(xGrid, 10)

scatter(xVals, yVals,legend=false, xlabel="Speed", ylabel = "Distance",
        xlims=(0, 30), ylims=(0,150))

plot!(xGrid, yEst)
```

### fitting $\lambda$

```julia
using RDatasets, DataFrames, Random, Statistics, LinearAlgebra
using MultivariateStats, LaTeXStrings, Plots; pyplot()
Random.seed!(0)

df = dataset("datasets", "cars")
n = size(df)[1]
df = df[shuffle(1:n),:]

K = 10
nG = Int(floor(n/K))
n = K*nG
println("Loosing $(size(df)[1] - n) observations.")


lamGrid = 0:100:30000

devSet(k) = collect(1+nG*(k-1):nG*k)
trainSet(k) = setdiff(1:n,devSet(k))

xTrain(k) = convert(Array{Float64,2},df[trainSet(k),[:Speed]])
xDev(k) = convert(Array{Float64,2},df[devSet(k),[ :Speed]])

yTrain(k) = convert(Array{Float64,1},df[trainSet(k),:Dist])
yDev(k) = convert(Array{Float64,1},df[devSet(k),:Dist])


errVals = zeros(length(lamGrid))
for (i,lam) in enumerate(lamGrid)
    errSamples = zeros(K)
    for k in 1:K
        beta = ridge(xTrain(k),yTrain(k),lam)
        errSamples[k] = norm([ones(nG) xDev(k)]*beta - yDev(k) )^2
    end
    errVals[i] = sqrt(mean(errSamples))
end

i = argmin(errVals)
bestLambda = lamGrid[i]


betaFinal = ridge(convert(Array{Float64,2},df[:,[:Speed]]),
                  convert(Array{Float64,1},df[:,:Dist]),bestLambda)

macro RR(x) return:(round.($x,digits = 3)) end
println("Found best lambda for regularization: ", bestLambda)
println("Beta estimate: ", @RR betaFinal)

plot(lamGrid, errVals,legend = false,
     xlabel = L"\lambda", ylabel = "Loss")
plot!([bestLambda,bestLambda],[0,100], c = :black, xlims = (0,3000),ylims = (0, 800))
```
