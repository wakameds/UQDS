# Cantor Set
Display the [Cantor set](https://en.wikipedia.org/wiki/Cantor_set) using Numpy and Matplotlib

First import the necessary libraries

```
import numpy as np
import matplotlib.pyplot as plt
```

Define the constants and parameters

```
depth = 5 #defines how many iterations to compute
unit_interval = [0, 1] #the line interval to divide
```

Define the recursive function to construct the Cantor set

```
def divide(interval, level=0, verbose=False):
    '''
    Divides a list of intervals into 3 and removes the middle segment
    '''
    if verbose:
        print("Interval:", interval)
    #plot
    plt.plot(interval, [level, level], lw=2, color="b")
    #generate line segments by creating 4 points along the current interval
    segments = np.linspace(interval[0], interval[1], 4)
    if level < depth:
        divide(segments[:2], level+1)
        divide(segments[2:], level+1)
```

Compute the set

```
divide(unit_interval)
```
