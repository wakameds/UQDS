### 1. Consider data points in the plane $(x_1, y_1), . . . ,(x_n, y_n)$, and assume you are searching for parameters of a function $f(x)$ such that,

$\sum_{i=1}^{n}(f(x_i)-y_i)^2$ is minimized.

####  (a) When $f(x) = β_0 + β_1x$, there are simple formulas for optimizing $β_0$ and $β_1$. See for example formulas (8.18) - (8.20) in [SWJ - draft version not for circulation available via BB]. Use the design matrix A as in formula (8.8) in [SWJ], and the pseudo-inverse $A^† = (A^T A)^{−1}A^T$ to derive the formulas in (8.18).

Let $A=\begin{bmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots & \vdots\\
1 & x_n
\end{bmatrix},\quad
A^T=\begin{bmatrix}
1 & 1 & \ldots & 1 \\
x_1 & x_2 & \ldots & x_n 
\end{bmatrix}
,\quad
\beta=\begin{bmatrix}
\beta_0\\
\beta_1
\end{bmatrix}
$

$f(x)=A\beta=y$

$A^TA\beta=A^Ty$

$\hat{\beta} = (A^TA)^{-1}A^T y = A^†y$

$A^TA=
\begin{bmatrix}
n & \sum_{i=1}^{n}{x_i}\\
\sum_{i=1}^{n}{x_i} & \sum_{i=1}^{n}{x_i}^2
\end{bmatrix}$

$(A^TA)^{-1}=\dfrac{1}{n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2}
\begin{bmatrix}
\sum_{i=1}^{n}{x_i}^2 & -\sum_{i=1}^{n}{x_i}\\
-\sum_{i=1}^{n}{x_i} & n
\end{bmatrix}$

so,

$\hat{\beta} = \begin{bmatrix}
\hat{\beta_0}\\
\hat{\beta_1}
\end{bmatrix}=
(A^TA)^{-1}A^T y = A^†y=\dfrac{1}{n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2}
\begin{bmatrix}
\sum_{i=1}^{n}{x_i}^2 & -\sum_{i=1}^{n}{x_i}\\
-\sum_{i=1}^{n}{x_i} & n
\end{bmatrix}
\begin{bmatrix}
1 & 1 & \ldots & 1 \\
x_1 & x_2 & \ldots & x_n 
\end{bmatrix}
\begin{bmatrix}
y_1\\
y_2\\
\vdots\\
y_n
\end{bmatrix}
$

$=\dfrac{1}{n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2}
\begin{bmatrix}
\sum_{i=1}^{n}{x_i}^2 & -\sum_{i=1}^{n}{x_i}\\
-\sum_{i=1}^{n}{x_i} & n
\end{bmatrix}
\begin{bmatrix}
\sum_{i=1}^{n}{y_i}\\
\sum_{i=1}^{n}{x_iy_i}
\end{bmatrix}$

$=\dfrac{1}{n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2}
\begin{bmatrix}
\sum_{i=1}^{n}{x_i}^2 \sum_{i=1}^{n}{y_i}-\sum_{i=1}^{n}{x_i}\sum_{i=1}^{n}{x_iy_i}\\
n\sum_{i=1}^{n}{x_iy_i}-\sum_{i=1}^{n}{x_i}\sum_{i=1}^{n}{y_i}
\end{bmatrix}
$

$\hat{\beta_0}=\dfrac{\sum_{i=1}^{n}{x_i}^2 \sum_{i=1}^{n}{y_i}-\sum_{i=1}^{n}{x_i}\sum_{i=1}^{n}{x_iy_i}}{n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2}, \quad
\hat{\beta_1} = \dfrac{n\sum_{i=1}^{n}{x_iy_i}-\sum_{i=1}^{n}{x_i}\sum_{i=1}^{n}{y_i}}{n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2}$

Also, $\bar{x}=\dfrac{\sum_{i=1}^{n}{x_i}}{n},\quad \bar{y}=\dfrac{\sum_{i=1}^{n}{y_i}}{n}$

$\bar{y}-\hat{\beta_1}\bar{x}=\dfrac{\sum_{i=1}^{n}{y_i}}{n}-\dfrac{\sum_{i=1}^{n}{x_i}}{n}\times \dfrac{n\sum_{i=1}^{n}{x_iy_i}-\sum_{i=1}^{n}{x_i}\sum_{i=1}^{n}{y_i}}{n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2}$

$=\dfrac{(n\sum_{i=1}^{n}{y_i}\sum_{i=1}^{n}{x_i}^2-\sum_{i=1}^{n}{y_i}(\sum_{i=1}^{n}{x_i})^2) -(n\sum_{i=1}^{n}{x_i}\sum_{i=1}^{n}{x_iy_i}-(\sum_{i=1}^{n}{x_i})^2\sum_{i=1}^{n}{y_i})}  
{n(n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2)}$

$=\dfrac{\sum_{i=1}^{n}{y_i}\sum_{i=1}^{n}{x_i}^2-\sum_{i=1}^{n}{x_i}\sum_{i=1}^{n}{x_iy_i}}  
{n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2} =\beta_0$

Hence,

$\hat{\beta_0}=\bar{y}-\hat{\beta_1}{\bar{x}}, \quad
\hat{\beta_1} = \dfrac{n\sum_{i=1}^{n}{x_iy_i}-\sum_{i=1}^{n}{x_i}\sum_{i=1}^{n}{y_i}}{n\sum_{i=1}^{n}{x_i}^2-(\sum_{i=1}^{n}{x_i})^2}$

#### (d) Consider Listing 8.8 in [SWJ]. Modify the code in this listing to find the estimates of $β_0$ and $β_1$, directly via the pseudo-inverse via: $\hat{β} = A^†y $ (where $\hat{β}$ is the vector of length $2$ and $y$ is the vectors of $y_1,...,y_n$). For each of the four datasets, find the point with the highest leverage (highest $H_{ii}$).

```julia
using RDatasets, DataFrames, GLM
df = dataset("datasets", "anscombe")
array = Matrix(df)

# x vector
X1 = array[:,1]
X2 = array[:,2]
X3 = array[:,3]
X4 = array[:,4]

V1 = ones(11)

#each model matrix A and sudo-
A1 = [V1 X1]
A2 = [V1 X2]
A3 = [V1 X3]
A4 = [V1 X4]

# pseudo inverse matrix
pseudoA1 = inv(A1'*A1)*A1' 
pseudoA2 = inv(A2'*A2)*A2' 
pseudoA3 = inv(A3'*A3)*A3' 
pseudoA4 = inv(A4'*A4)*A4' 

# each y vector
Y1 = array[:,5]
Y2 = array[:,6]
Y3 = array[:,7]
Y4 = array[:,8]

# beta
beta1 = pseudoA1*Y1
beta2 = pseudoA2*Y2
beta3 = pseudoA3*Y3
beta4 = pseudoA4*Y4

println("Model 1. beta_0 and beta_1: ", beta1)
println("Model 2. beta_0 and beta_1: ", beta2)
println("Model 3. beta_0 and beta_1: ", beta3)
println("Model 4. beta_0 and beta_1: ", beta4)
```

```julia
function Hatmax(H)
    n,m = size(H)
    max = 0
    list = []
    for i in 1:n
        for j in 1:m
            if i == j
                append!(list, H[i,j])
            end
        end
    end
    list
end


function selecthighest(H)
    maxval = maximum(H)
    positions = [i for (i, x) in enumerate(H) if x == maxval]
    println("The highest Hii is: ", maxval, "\t i is: ", positions)
end   
```

```julia
Hat1 = A1*pseudoA1    
Hat2 = A2*pseudoA2
Hat3 = A3*pseudoA3
Hat4 = A4*pseudoA4


selecthighest(Hatmax(Hat1))
selecthighest(Hatmax(Hat2))
selecthighest(Hatmax(Hat3))
selecthighest(Hatmax(Hat4))
```

#### (e) Assume now that you wish to fit $f(x) = β_0 + β_1x + β_2x^2$. Here the design matrix $A$ is $n × 3$. Repeat the fitting of Listing 8.8 using this type of model. Display your fit model graphically.

```julia
using RDatasets, DataFrames, GLM
df = dataset("datasets", "anscombe")
array = Matrix(df)

# x vector
X1 = array[:,1]
X2 = array[:,2]
X3 = array[:,3]
X4 = array[:,4]


#each model matrix A and sudo-
A1 = [ones(11) X1 X1.^2]
A2 = [ones(11) X1 X2.^2]
A3 = [ones(11) X1 X3.^2]
A4 = [ones(11) X1 X4.^2]

# each y vector
Y1 = array[:,5]
Y2 = array[:,6]
Y3 = array[:,7]
Y4 = array[:,8]

newBetaHat1 = inv(A1'A1)*A1'*Y1
newBetaHat2 = inv(A2'A2)*A2'*Y2
newBetaHat3 = inv(A3'A3)*A3'*Y3
newBetaHat4 = inv(A4'A4)*A4'*Y4
```

```julia
newPred1(x) = newBetaHat1'*[1, x, x^2]
newPred2(x) = newBetaHat2'*[1, x, x^2]
newPred3(x) = newBetaHat3'*[1, x, x^2]
newPred4(x) = newBetaHat4'*[1, x, x^2]
```

```julia
using LinearAlgebra, Statistics, Plots, Random, Distributions, StatsPlots
xPltPoints=0:20
scatter(X1,Y1, legend=true, xlabel="X", ylabel = "Y", xlim=(0,20),ylim=(0,20),label="Y1")
plot!(xPltPoints, newPred1.(xPltPoints), label="Model1") 
```

```julia
using LinearAlgebra, Statistics, Plots, Random, Distributions, StatsPlots
xPltPoints=0:20
scatter(X2,Y2,legend=true, xlabel="X", ylabel = "Y", xlim=(0,20),ylim=(0,20), label="Y2")
plot!(xPltPoints, newPred2.(xPltPoints), label="Model2") 
```

```julia
using LinearAlgebra, Statistics, Plots, Random, Distributions, StatsPlots
xPltPoints=0:20
scatter(X3,Y3,legend=true, xlabel="X", ylabel = "Y", xlim=(0,20),ylim=(0,20), label="Y3")
plot!(xPltPoints, newPred3.(xPltPoints), label="Model3") 
```

```julia
using LinearAlgebra, Statistics, Plots, Random, Distributions, StatsPlots
xPltPoints=0:20
scatter(X4,Y4,legend=true, xlabel="X", ylabel = "Y", xlim=(0,20),ylim=(0,20), label="Y4")
plot!(xPltPoints, newPred4.(xPltPoints), label="Model4") 
```

#### (a) Present the companion matrix for the polynomial $p(u) = (2 − u)(7 − u)(9 − u)$.

$a_0=-126, \quad a_1=95, \quad a_2=-18$

0 & 0 & 1 \\

\end{bmatrix}

#### (b) Compute the characteristic polynomial of the companion matrix for this polynomial. Show that it equals $p(u)$.

$C_p - uI = \begin{bmatrix}
-u & 1 & 0\\
0 & -u & 1\\
126 & -95 & 18 - u
\end{bmatrix}$

$det(C_p - uI)= -u((-u)(18-u)-(1)(-95))-1((0)(18-u)-(1)(126))+0((0)(-95)-(1)(126))$

$= 126-95u+18u^2-u^3$

$= (2-u)(7-u)(9-u)$

$= p(u)$
